```python
#create feature and label(text, target)

X = train_data['text']
y = train_data['target']
```

```python
X[:2], y[:2]
```

```python
# Important step: train_test_split

from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.1, random_state=42)
```

```python
#Text Vectorization

#we can't just use strings in our model, we need to convert it to some numerical form.
#the first step is to use text vectorization.

from tensorflow.keras.layers import TextVectorization

max_tokens = 10000
max_seq_length = 15

vectorizer = TextVectorization(
    max_tokens = max_tokens,
    split='whitespace',
    standardize = 'lower_and_strip_punctuation',
    output_mode = 'int',
    output_sequence_length = max_seq_length
)

vectorizer.adapt(X_train)
```

```python
#Embedding layer:
#we can use the embedding layer to make our numerical form learnable.

from tensorflow.keras.layers import Embedding

output_dim = 64

embedding_layer = Embedding(
    input_dim = max_tokens,
    output_dim = output_dim,
    embeddings_initializer = 'uniform'
)
```

```python
#Building, Compiling and training model:

from tensorflow.keras.layers import Input, Dense, LSTM, GlobalAveragePooling2D, Flatten
from tensorflow.keras import Model
```

```python
input_layer = Input(shape=(1,), dtype='string')
x = vectorizer(input_layer)
x = embedding_layer(x)
x = LSTM(64)(x)
x = Flatten()(x)
x = Dense(16, activation='relu')(x)
output_layer = Dense(1, activation='sigmoid')(x)
lstm_model = Model(input_layer, output_layer)
```

```python
#Compiling(Adam optimizer, binary_crossentropy loss, accuracy metric)

lstm_model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.05),
    loss = tf.keras.losses.binary_crossentropy,
    metrics = ['accuracy']
)
```

```python
#fitting model with train data

lstm_history = lstm_model.fit(
    X_train,
    y_train,
    epochs=5,
    validation_data=(X_val, y_val)
)
```

```python
#Evaluating Model 📉

lstm_model.evaluate(X_val, y_val)
```

```python
def plot_loss_accuracy_curves(history):
  print(pd.DataFrame(history.history))
  number_of_epochs = len(history.history['loss'])
  plt.figure()
  plt.plot(range(1, number_of_epochs+1), history.history['loss'], label='loss')
  plt.plot(range(1, number_of_epochs+1), history.history['val_loss'], label='validation loss')
  plt.legend();

  plt.figure()
  plt.plot(range(1, number_of_epochs+1), history.history['accuracy'], label='accuracy')
  plt.plot(range(1, number_of_epochs+1), history.history['val_accuracy'], label='validation accuracy')
  plt.legend();
```

```python
plot_loss_accuracy_curves(lstm_history)
```

```python
#Using GRU model, for our nlp problem

#we do the same things we did in LSTM modelling

from tensorflow.keras.layers import GRU

input_layer_1 = Input(shape=(1,), dtype='string')
x_1 = vectorizer(input_layer_1)
x_1 = embedding_layer(x_1)
x_1 = GRU(64)(x_1)
x_1 = Flatten()(x_1)
x_1 = Dense(16, activation='relu')(x_1)
output_layer_1 = Dense(1, activation='sigmoid')(x_1)
gru_model = Model(input_layer_1, output_layer_1)
```

```python
gru_model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.binary_crossentropy,
    metrics=[
        'accuracy'
    ]
)
```

```python
gru_history = gru_model.fit(
    X_train,
    y_train,
    epochs=10,
    validation_data=(X_val, y_val)
)
```

```python
gru_model.evaluate(X_val, y_val)
```

```python
plot_loss_accuracy_curves(gru_history)
```

```python
y_preds_lstm = tf.round(lstm_model.predict(X_val))
print(y_preds_lstm)
y_preds_gru = tf.round(gru_model.predict(X_val))
print(y_preds_gru)
```

```python
from sklearn.metrics import classification_report

print("LSTM model classification report \n")
print(classification_report(y_val, y_preds_lstm))

print("\nGRU model classification report \n")
print(classification_report(y_val, y_preds_gru))
```

```python
#Applying SVC to the problem

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC

vec_1 = TfidfVectorizer().fit(X_train)
X_train_vec = vec_1.transform(X_train)
X_val_vec = vec_1.transform(X_val)

svc_model = SVC(
    C=0.5,
    kernel='rbf'
)

svc_model.fit(X_train_vec, y_train)
```

```python
y_preds = svc_model.predict(X_val_vec)
print(classification_report(y_val, y_preds))
```

```python
#Transfer learning, reduce overfitting, better accuracy, better model!

import tensorflow_hub as hub

# for use model i want to use Sequential Api
use_model = tf.keras.Sequential()
# adding layers
use_model.add(
    hub.KerasLayer(
        "https://tfhub.dev/google/universal-sentence-encoder/4",
         trainable=False,
         input_shape=[],
         dtype='string'
    )
)

use_model.add(tf.keras.layers.Dense(32, activation='relu'))
use_model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

```

```python
# compile use model
use_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
# fit the model 
use_model.fit(
    X_train,
    y_train,
    epochs=10,
    validation_data=(X_val, y_val)
)
```
